```python
import pinecone
from pinecone.core.client.model import IndexSpec

# Initialize Pinecone client
client = pinecone.Pinecone(api_key=PINECONE_API_KEY, environment=PINECONE_ENVIRONMENT)

# Define the index name and specifications
index_name = "gemini-rag-index"

# Check if the index exists; if not, create it
if index_name not in client.list_indexes():
    index_spec = IndexSpec(name=index_name, dimension=768, metric="cosine")
    client.create_index(spec=index_spec)

# Connect to the index
index = client.Index(index_name)

print(f"Connected to Pinecone index: {index_name}")
```
